```python
import numpy as np


# Maksimum Generasi
N = 8
#Batas atas dan batas bawah
ra = 5
rb = -5
#Batas atas dan bawah pada X dan Y
ra_x = 5
rb_x = -5
ra_y = 5
rb_y = -5

#jumlah kromosom in population
jml_kromosom = 8
#Membuat populasi Random
populasi_awal = np.random.randint(10, size = (jml_kromosom,10))
temp_decode = [1/10,1/100,1/1000,1/10000,1/100000]
```

```python
#Dekode Kromosom
def decodeKromosom(decode):
  decode_x = rb_x + ((ra_x - rb_x)/(9*sum(temp_decode))) * ((decode[0]*temp_decode[0]) + (decode[1]*temp_decode[1]) +
             (decode[2]*temp_decode[2]) + (decode[3]*temp_decode[3])+(decode[4]*temp_decode[4]))
  decode_y = rb_y + ((ra_y - rb_y)/(9*sum(temp_decode))) * ((decode[5]*temp_decode[0]) + (decode[6]*temp_decode[1]) +
             (decode[7]*temp_decode[2]) + (decode[8]*temp_decode[3])+(decode[9]*temp_decode[4]))
  return decode_x, decode_y
```

```python
#Menghitung fungsi fitness
def FungsiFitness(x,y):
  a = 1/100
  h = ((np.cos(x) + np.sin(y))**2)/x**2+y**2
  fitness = 1/h+a
  return fitness
```

```python
#Pencarian Index Parent
def find_orangtua(nilai_fitness) :
  jumlah = sum(nilai_fitness)
  r = np.random.random()
  index = 0
  while r > 0 and index < 9 :
    r = r - (nilai_fitness[index]/jumlah)
    index += 1
  return index
```

```python
#Crossover populasi untuk menghasilkan turunan baru dari parent
def crossover(new_populasi, nilai_fitness) :
  for i in range(jml_kromosom//2) :
    p1 = find_orangtua(nilai_fitness)
    p2 = find_orangtua(nilai_fitness)
    if (np.random.random() < 8/10):
      tipot = np.random.randint(8)
      child1 = np.array(list(populasi_awal[p1][tipot:]) + list(populasi_awal[p2][:tipot]))
      child2 = np.array(list(populasi_awal[p2][tipot:]) + list(populasi_awal[p1][:tipot]))
      new_populasi.append(child1)
      new_populasi.append(child2)
    else:
      new_populasi.append(populasi_awal[p1])
      new_populasi.append(populasi_awal[p2])
```

```python
#Mutasi populasi representasi integer (Memilih nilai secara acak)
def mutasi(new_populasi) :
  for i in range(jml_kromosom):
    for j in range(10):
      if (np.random.random() < 0.1):
        new_populasi[i][j] = np.random.randint(9)
```

```python
#Cari index kromosom dengan fitness yang terbesar
def index_best(nilai_fitness):
  idx = 0
  for i in range(len(nilai_fitness)):
    if (nilai_fitness[i] > nilai_fitness[idx]):
      idx = i
  return idx
```

```python
#Cari index kromosom dengan fitness yang terkecil
def index_bad(nilai_fitness):
  idx = 0
  for i in range(len(nilai_fitness)):
    if (nilai_fitness[i] < nilai_fitness[idx]):
      idx = i
  return idx
```

```python
#Main Program

populasi = populasi_awal
best_fitness = []
bad_fitness = []

#Loop Generasi
for gen in range(N):
  #Populasi awal
  if gen == 0:
    print("Populasi awal :")
    for i in range(jml_kromosom):
      print("Kromosom", i+1, ":", populasi_awal[i])
    print()

  #output populasi Tiap Generasi
  if gen+1 != N:
    print("====================== generasi ke ", gen+1, "=======================")
    nilai_fitness = []
    new_populasi = []

    #output populasi dekode dan fitness
    for i in range(jml_kromosom):
      x,y = decodeKromosom(populasi[i])
      nilai_fitness.append(FungsiFitness(x,y))
      print("Kromosom", i+1, ":", populasi[i])
      print("Dekode Titik X :", x)
      print("Dekode Titik Y : ", y)
      print("Fitness :", FungsiFitness(x,y))

    bestkrom = populasi[index_best(nilai_fitness)]

    #Crossover dan mutasi populasi
    crossover(new_populasi, nilai_fitness)
    mutasi(new_populasi)

    #Elitisme
    nilai_childfitness = []
    for j in range(jml_kromosom):
      x,y = decodeKromosom(new_populasi[j])
      nilai_childfitness.append(FungsiFitness(x,y))
    badkrom = new_populasi[index_bad(nilai_childfitness)]
    xbest , ybest = decodeKromosom(bestkrom)
    xbad , ybad = decodeKromosom(badkrom)

    if FungsiFitness(xbest, ybest) > FungsiFitness(xbad, ybad):
      new_populasi[index_best(nilai_childfitness)] = bestkrom

    best_fitness.append(FungsiFitness(xbest,ybest))
    bad_fitness.append(FungsiFitness(xbad, ybad))

    #output populasi baru
    print("Populasi Baru generasi ke", gen+2 , ":")
    for i in range(jml_kromosom):
      print("Kromosom", i+1, ":", new_populasi[i])
    print()

    populasi = new_populasi
  else :
    print("=====================generasi terakhir/ke", gen+1,"===================")
    nilai_fitness = []
    for i in range(jml_kromosom):
      x,y = decodeKromosom(populasi[i])
      nilai_fitness.append(FungsiFitness(x,y))
      print("Kromosom", i+1 , ":", populasi[i])
      print("Dekode Titik X :", x)
      print("Dekode Titik Y : ",y)
      print("Fitness :", FungsiFitness(x,y))
      print()
```
